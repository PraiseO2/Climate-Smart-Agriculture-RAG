###  Load PDFs

```python
from langchain_community.document_loaders import PyPDFLoader

all_documents = []
for pdf_file in sorted(os.listdir(PDF_FOLDER)):
    if not pdf_file.endswith('.pdf'):
        continue
    try:
        loader = PyPDFLoader(os.path.join(PDF_FOLDER, pdf_file))
        docs = loader.load()
        all_documents.extend(docs)
        print(f"✅ {pdf_file} — {len(docs)} pages")
    except Exception as e:
        print(f"❌ {pdf_file} — {e}")

print(f"\n Total pages: {len(all_documents)}")
```

### Chunk Documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500, chunk_overlap=50,
    separators=["\n\n", "\n", ".", " ", ""]
)
chunks = splitter.split_documents(all_documents)
print(f"✅ {len(chunks):,} chunks created from {len(all_documents):,} pages")
```

### Load Embedding Model

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={'device': 'cuda'},
    encode_kwargs={'normalize_embeddings': True}
)
test = embeddings.embed_query("climate smart agriculture")
print(f"✅ Embeddings loaded — {len(test)} dimensions")
```

### Build FAISS Vector Store

```python
from langchain_community.vectorstores import FAISS
import time

VECTORSTORE_PATH = '/content/drive/MyDrive/Agriculture_Project/Food_Climate_RAG/Climate_Food_RAG_Project/vectorstore_faiss_new'

print(f"Building vector store from {len(chunks):,} chunks...")
start = time.time()
vectorstore = FAISS.from_documents(chunks, embeddings)
vectorstore.save_local(VECTORSTORE_PATH)
print(f"✅ Done in {(time.time()-start)/60:.1f} mins — {vectorstore.index.ntotal:,} vectors saved")
```

### Test Retriever

```python
vectorstore = FAISS.load_local(
    VECTORSTORE_PATH, embeddings,
    allow_dangerous_deserialization=True
)
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})

results = retriever.invoke("What is climate smart agriculture?")
for i, doc in enumerate(results, 1):
    source = doc.metadata.get('source','').split('/')[-1]
    page = doc.metadata.get('page','N/A')
    print(f"\n📄 Result {i} — {source} | Page {page}")
    print(f"   {doc.page_content[:200]}...")
```

### Load LLM

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

MODEL_NAME = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_NAME, torch_dtype=torch.float16, device_map="auto"
)
print(f"✅ Model loaded on {next(model.parameters()).device}")
```

### Prompt Template

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template="""You are an expert agricultural AI assistant for climate-smart agriculture and food security in sub-Saharan Africa.

Answer ONLY from the context below. If insufficient, say "I don't have enough information in my knowledge base."
Be specific and practical for both farmers and researchers.

CONTEXT: {context}

QUESTION: {question}

ANSWER:""",
    input_variables=["context", "question"]
)
print("✅ Prompt template ready")
```

### Build RAG Chain

```python
from langchain_core.language_models.llms import LLM
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from typing import Any, List, Optional
import torch

class FlanT5LLM(LLM):
    max_new_tokens: int = 512
    temperature: float = 0.7
    repetition_penalty: float = 1.3

    @property
    def _llm_type(self): return "flan-t5-large"

    def _call(self, prompt: str, stop=None, **kwargs) -> str:
        inputs = tokenizer(
            prompt, return_tensors="pt",
            max_length=512, truncation=True, padding=True
        ).to(model.device)
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=self.max_new_tokens,
                temperature=self.temperature,
                do_sample=True,
                repetition_penalty=self.repetition_penalty,
                no_repeat_ngram_size=3,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
            )
        return tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

    @property
    def _identifying_params(self):
        return {"model_name": "google/flan-t5-large"}

def format_docs(docs):
    return "\n\n".join([
        f"[{doc.metadata.get('source','').split('/')[-1]}, p.{doc.metadata.get('page','?')}]\n{doc.page_content[:300]}"
        for doc in docs
    ])

llm = FlanT5LLM()

prompt = PromptTemplate(
    template="""You are an agricultural expert. Using the context below, write a detailed
and practical answer to the question. Include specific examples, explain why they work,
and mention which farmers or regions benefit most.

Context:
{context}

Question: {question}

Detailed answer:""",
    input_variables=["context", "question"]
)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt | llm | StrOutputParser()
)
print("✅ Final RAG chain ready")
```

### First Real Test

```python
import time

questions = [
    "How can smallholder farmers adapt to drought in Africa?",
    "What is climate smart agriculture?",
    "What crops are drought tolerant in sub-Saharan Africa?"
]

for q in questions:
    print(f"{'='*60}")
    print(f"❓ {q}\n")
    start = time.time()
    answer = rag_chain.invoke(q)
    print(f"🤖 ANSWER:\n{answer}")
    print(f"\n⏱️  {time.time()-start:.1f}s")

print(f"\n{'='*60}")
```

### Gradio Interface

```python
import gradio as gr

def answer_question(question):
    if not question.strip():
        return "Please enter a question."
    try:
        return rag_chain.invoke(question)
    except Exception as e:
        return f"Error: {str(e)}"

with gr.Blocks(title="Climate-Smart Agriculture AI") as demo:

    gr.Markdown("""
    # 🌾 Climate-Smart Agriculture AI Assistant
    Powered by RAG — answers grounded in FAO, IPCC and CGIAR research

    *Built by Praise Adetifa — B.Agric, OAU Ile-Ife*
    """)

    with gr.Row():
        with gr.Column(scale=1):
            question_input = gr.Textbox(
                label="Your Question",
                placeholder="e.g. How can farmers adapt to drought?",
                lines=3
            )
            submit_btn = gr.Button("Ask →", variant="primary")
            gr.Examples(
                examples=[
                    ["What is climate smart agriculture?"],
                    ["How can smallholder farmers adapt to drought in Africa?"],
                    ["What crops are drought tolerant in sub-Saharan Africa?"],
                    ["How does climate change affect food security?"],
                    ["What is the role of agroforestry in climate adaptation?"],
                ],
                inputs=question_input
            )

        with gr.Column(scale=1):
            answer_output = gr.Textbox(
                label="Answer",
                lines=10,
                show_copy_button=True
            )

    submit_btn.click(fn=answer_question,
                     inputs=question_input,
                     outputs=answer_output)
    question_input.submit(fn=answer_question,
                          inputs=question_input,
                          outputs=answer_output)

demo.launch(share=True)
```
